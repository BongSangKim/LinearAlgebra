### Contents (1.5 - 1.6)

* LU decompoisition
  * Can All $n\times n$ matrices be decomposed into $LU$?
  * Example:
    $$A = \begin{bmatrix} 1 & 1 & 1 \\ 1 & 1 & 3 \\ 2 & 5 & 8 \end{bmatrix}$$
  * Definition: permutation matrices
  * Example:
    $$P_{23}E_{31}(-2)E_{21}(-1)A = \begin{bmatrix} 1 & 1 & 1 \\ 0 & 3 & 6 \\ 0 & 0 & 2 \end{bmatrix}$$
  * Can we switch rows at the first place?
  * Theorem: For any $n\times n$ nonsingular matrix $A$, there exists a permutation matrix $P$ such that $PA$ is decomposed into $PA = LU$.
  * **Q** How to find $P$?
  * LDU decomposition
  
* Gauss-Jordan elimination
  * Definition: reduced row echelon from (RREF)
    * Every pivot (the first nonzero entry of a row) is $1$.
    * 0 entries above and below the pivots.
    * The pivor columns (columns containing the pivots) are arranged in increasing order.
  * Let $A$ be a nonsingular matrix and $[I|B]$ be the RREF of $[A|I]$. Then $B$ is the inverse of $A$.
  * Example: find the inverse of 
    $$A = \begin{bmatrix} 1 & 1 & 1 \\ 1 & 1 & 3 \\ 2 & 5 & 8 \end{bmatrix}$$
  * **Q** Is the inverse of a (nonsingular) lower triangular matrix also a lower triangluar? How about upper triangular matrices?
  * Uniqueness of LDU decomposition
  
  
* Transpose and symmetric matrices
  * Property : $(AB)^T = B^TA^T$, $(A^{-1})^T = (A^T)^{-1}$
  * $LDU$ factorization of symmetric matrix: $LDL^T$
  * **Q** Re-visit the question: if a left (or right) inverse exists, then a right (or left) inverse exists.
